# SBA - Data Analysis and Visualization with Python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. read data from file
dataset = pd.read_csv('woocommerce-product-export.csv')
dataset
```

```python
# 2. Summary of the columns 
dataset.info()
```

```python
# 3. Summary of statistics 
dataset.describe()
```

```python
# 4. First five rows by default 
print(dataset.head())
```

```python
# 5. last five rows by default
print(dataset.tail())
```

```python
# 6. Print  the "total_profit" and “month_number” columns only
print(dataset[['total_profit', 'month_number']])
```

```python
# 7. Read the total profit of all months and show it using the Bar plot.
plt.figure(figsize=(8, 5))
plt.bar(dataset['month_number'], dataset['total_profit'])
plt.xlabel('Month number')
plt.ylabel('Total profit')
plt.yticks(range(100000, 500001, 100000))
plt.title('Company profit per month')
plt.grid(linestyle='--')
plt.show() 
```

```python
# 8. Read the total profit of all of the months, and show the line plot

plt.figure(figsize=(8, 5))
plt.plot(
    dataset['month_number'], 
    dataset['total_profit'], 
    label='Profit data of last year', 
    linestyle='--', 
    linewidth=3, 
    c='red', 
    marker='o', 
    markerfacecolor='black'
)

plt.xlabel('Month Number')
plt.ylabel('Total profit')
plt.yticks(range(100000, 500001, 100000))
plt.title('Company profit per month')
plt.legend(loc='lower right')
plt.show()
```

```python
# 9. Print all of the product sales data and show it using a multi-line plot

print(dataset)
plt.figure(figsize=(10, 6))

products = [
    'facecream',
    'facewash',
    'toothpaste',
    'bathingsoap',
    'shampoo',
    'moisturizer'
]

print('Multi-Line Plot:')

for product in products:
    plt.plot(
        dataset['month_number'],
        dataset[product],
        marker='o',
        linewidth=2,
        label=f'{product.capitalize()} Sales Data'  # .capitalize() capitalizes the first letter of the product name on the legend label 
    )
plt.xlabel('Month Number')
plt.xticks(range(1, 13))
plt.ylabel('Sales units in number')
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.title('Sales data')
plt.legend(loc='upper left')

plt.show()
```

```python
# 10. Read “bathingsoap” sales data for each month and show it using a scatter plot.

plt.figure(figsize=(8, 5))
plt.scatter(dataset['month_number'], dataset['bathingsoap'], label='Bathing soap Sales data')
plt.xlabel('Month Number')
plt.xticks(range(1, 13))
plt.ylabel('Number of units Sold')
plt.title('bathingsoap Sales data')
plt.legend()
plt.grid(linestyle='--')
plt.show()
```
